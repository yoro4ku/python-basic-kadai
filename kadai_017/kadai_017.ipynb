```python
class Human:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def check_adult(self) ->str:
        if self.age >= 20:
            return "大人です。"
        else:
            return "大人ではありません。"

# インスタンス作成
users = [
    Human("侍太郎",36),
    Human("侍一郎",33),
    Human("侍二郎",19),
    Human("侍三郎",25),
    Human("侍四郎",12)
]

for user in users:
    print(f"名前:{user.name} 年齢:{user.age} 判定:{user.check_adult()}")

```
